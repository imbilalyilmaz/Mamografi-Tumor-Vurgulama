```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./images/mamografi.jpg', cv.IMREAD_GRAYSCALE)
```

## Histograma EÅŸitleme

```python
hist_eq = cv.equalizeHist(image)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Orijinal GÃ¶rÃ¼ntÃ¼')
plt.subplot(1, 2, 2), plt.imshow(hist_eq, cmap='gray'), plt.title('Histogram EÅŸitleme')
plt.show()
```

## CLAHE

```python
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(image)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Orijinal GÃ¶rÃ¼ntÃ¼')
plt.subplot(1, 2, 2), plt.imshow(clahe_img, cmap='gray'), plt.title('CLAHE')
plt.show()
```

## Canny

```python
canny_hist = cv.Canny(hist_eq, 125, 175)
canny_clahe = cv.Canny(clahe_img, 125, 175)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(canny_hist, cmap='gray'), plt.title('Histogram EÅŸitleme Ä°le Canny')
plt.subplot(1, 2, 2), plt.imshow(canny_clahe, cmap='gray'), plt.title('CLAHE ile Canny')
plt.show()
```

## GaussianBlur ve Threshold

```python
blur  = cv.GaussianBlur(clahe_img, (5,5), cv.BORDER_DEFAULT)

ret, thresh = cv.threshold(blur, 125, 255, cv.THRESH_BINARY)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1), plt.imshow(blur, cmap='gray'), plt.title('BulanÄ±klaÅŸtÄ±rÄ±lmÄ±ÅŸ GÃ¶rÃ¼ntÃ¼')
plt.subplot(1, 2, 2), plt.imshow(thresh, cmap='gray'), plt.title('Threshold')
plt.show()
```
